```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager

import pandas as pd
import matplotlib.pyplot as plt
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

url = 'https://data-class-mars-challenge.s3.amazonaws.com/Mars/index.html'
browser.visit(url)
html = browser.html
mars_soup = soup(html, 'html.parser')
```

```python
table = mars_soup.find('table', class_='table')

list = []

rows = table.find_all('tr')

for row in rows:
    table_heading = row.find('th')
    table_data = row.find('tr', class_='data-row')
    list.append(table_data)

#for row in rows:
   # row_heading = row.find('th').text
   # row_data = row.find('td').text.strip()
   # mars_facts[row_heading] = row_data
```

```python
min_temp = df.groupby(by=['month'])['min_temp'].mean()

min_temp_df = pd.DataFrame(min_temp)

min_temp_chart = min_temp_df.plot.bar(rot=0)

min_temp_chart

#Lowest Tmepereature: March Highest Temperature: August
```

```python
avg_pressure = df.groupby(by=['month'])['pressure'].mean()

avg_pressure_df = pd.DataFrame(avg_pressure)

chart = avg_pressure_df.plot.bar(rot=0)

chart
```

```python
x = df['sol']
y = df['min_temp']

plt.plot(x,y)
plt.xlabel("Earth days on Mars")
plt.ylabel("Minimum avg temperature")
plt.title("Minimum average temperature per Earth days on Mars")

plt.show()

# Around 670 days in a martian year (min to min)
```
